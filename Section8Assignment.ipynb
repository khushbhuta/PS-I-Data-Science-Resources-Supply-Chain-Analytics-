```python
countries = ['United Kingdom','EIRE','Canada','Denmark']
twentyeleven_uk = twentyeleven[twentyeleven.Country.isin(countries)]
```

```python
twentyeleven_uk['InvoiceDate'] = pd.to_datetime(twentyeleven_uk['InvoiceDate']) 
twentyeleven_uk['date'] = twentyeleven_uk['InvoiceDate'].dt.strftime('%Y-%m-%d')
twentyeleven_uk['date'] = pd.to_datetime(twentyeleven_uk['date'])
```

```python
sales_per_day = twentyeleven_uk.groupby(['Country','date']).agg(total_sales=('Quantity','sum')).reset_index()
```

```python
sales_per_day.plot()
```
