```python
retail = pd.read_csv("C:\\Users\\Khush Bhuta\\Desktop\\python_workspace\\data\\online_retail2 (1).csv")
retail = retail.drop_duplicates()
retail = retail.dropna(axis = 0, how = 'any')
retail.head()
```

```python
retail.Country.value_counts()
```

```python
ire_france = retail[retail.Country.isin(['EIRE','France'])]
```

```python
sns.countplot(x='Country',data=ire_france)
```

```python
ire_france['InvoiceDate'] = pd.to_datetime(ire_france['InvoiceDate'])
ire_france['dayofweek'] = ire_france['InvoiceDate'].dt.dayofweek
```

```python
ire_france['dayofweek']
```

```python
week_dict = {0:'Monday',1:'Tuesday',2:'Wednesday',3:'Thursday',4:'Friday',5:'Saturday',6:'Sunday'}
ire_france['dayofweek'] = ire_france['dayofweek'].map(week_dict)
```

```python
ire_france['dayofweek'].head()
```

```python
grouped = ire_france.groupby(['Country','dayofweek']).agg(total_quantity=('Quantity',np.sum)).reset_index()
grouped = grouped[grouped.total_quantity>=0]
```

```python
grouped
```

```python
sns.barplot(x='dayofweek',y='total_quantity',data = grouped,hue='Country')
```
