```python
data = data.drop_duplicates()
data = data.dropna(axis = 0)
```

```python
grouped_sku = data.groupby('Description').agg(total_sales = ('Quantity','sum'),
                                             total_revenue = ('revenue','sum')).reset_index()
```

```python
product_mix = inv.productmix(grouped_sku['Description'],grouped_sku['total_sales'],grouped_sku['total_revenue'])
```

```python
product_mix.product_mix.value_counts()
```

```python
grouped_sku_country = data.groupby(['Description','Country']).agg(total_sales = ('Quantity','sum'),
                                             total_revenue = ('revenue','sum')).reset_index()
grouped_sku_country
```

```python
product_mix_store_level = inv.productmix_storelevel(grouped_sku_country['Description'],
                                                   grouped_sku_country['total_sales'],
                                                   grouped_sku_country['total_revenue'],
                                                   grouped_sku_country['Country'])
```
