```python
car_pricing = cars_data[['cars name','Price']]
car_pricing
```

```python
def pricing_category(price):
    if(price<20000):
        return("Budget Car")
    elif((price>=20000) & (price<35000)):
        return("Suitable Car")
    else:
        return("Expensive Car")
```

```python
car_pricing['category'] = 'NA'
```

```python
car_pricing
```

```python
car_pricing = car_pricing.drop(columns = ['Type'])
```

```python
car_pricing
```

```python
for i in range(car_pricing.shape[0]):
    car_pricing.loc[i,'category'] = pricing_category(car_pricing.loc[i,'Price'])

car_pricing
```
